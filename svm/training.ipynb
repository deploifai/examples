```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
import os, pathlib
data_directory = os.environ.get("USER_EXPERIMENT_DATA_DIRECTORY", "/mnt/data/")
data = pd.read_csv(pathlib.Path(f"{data_directory}/data/container1/emails.csv"))
```

```python
data.head()
```

```python
data.describe()
```

```python
X = data.iloc[:, 1:3001]
X
```

```python
Y = data.iloc[:, -1]
Y
```

```python
X_occurrences = X.sum()
X_spam_score = {}
for col in X.columns:
  A = X[col].to_numpy().reshape(1,-1)
  B = Y.to_numpy().reshape(1, -1)
  value = np.matmul(A, B.T)[0][0]
  X_spam_score[col] = value
```

```python
X_occurrences
```

```python
X_spam_score
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(200,100))

x_points, y_points = [], []

for col in X.columns:
  x, y = X_occurrences[col], X_spam_score[col]/X_occurrences[col]
  if x < 10000:
    x_points.append(x)
    y_points.append(y)
    ax.annotate(col, (y, x))

ax.scatter(y_points, x_points)

ax.grid(True)
fig.tight_layout()

plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
svm = SVC(C=1.0, kernel='rbf', gamma='auto')
model = svm.fit(X_train, Y_train)
```

```python
acc = model.score(X_test, Y_test)
```

```python
metrics = {"accuracy": acc}
metrics
```

```python
artifacts_directory = os.environ.get("USER_EXPERIMENT_ARTIFACTS_PATH", "artifacts")

metrics_filepath = os.path.join(artifacts_directory, "metrics.json")

import json

with open(metrics_filepath, 'w') as metrics_file:
    json.dump(metrics, metrics_file, indent=2)
```

```python
import pickle

pkl_filepath = os.path.join(artifacts_directory, "model.pkl")

with open(pkl_filepath, 'wb') as model_file:
    pickle.dump(model, model_file)
```
