```python
import pandas as pd
import numpy as np
import sklearn
import pickle
```

```python
#importing the data
df = pd.read_csv("car_data.csv")
df.head()
```

```python
df.shape
```

```python
df=df.drop(columns="Car_Name")
df
```

```python
df['Age'] = 2023 - df['Year'] #creating a column for the age of the car
df.drop('Year',axis=1,inplace = True)
```

```python
df.head()
```

```python
df.rename(columns={"Selling_Price": "Selling_Price(lakhs)"}, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.get_dummies(data = df, drop_first=True)
df 
```

```python
y = df['Selling_Price(lakhs)'] #target
X = df.drop('Selling_Price(lakhs)',axis=1) #features
#splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import scale



rfr = RandomForestRegressor()

X=scale(X)
y=scale(y)

rfr.fit(X_train,y_train)
  
# R2 score of train set
y_pred_train = rfr.predict(X_train)
R2_train_model = r2_score(y_train,y_pred_train)

# R2 score of test set
y_pred_test = rfr.predict(X_test)
R2_test_model = r2_score(y_test,y_pred_test)



```

```python
print("r2 score for training set: ", R2_train_model)
print("r2 score for test set: ", R2_test_model )
```

```python
pickle.dump(rfr, open("RandomForestRegressor.pkl", "wb"))
```
